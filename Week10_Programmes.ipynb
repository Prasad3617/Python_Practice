## Stacking Numpy Arrays

```python
img_h=np.hstack((img,img))       #Horizontally stacking image
```

```python
cv2_imshow(img_h)
```

```python
img_v=np.vstack((img,img))
```

## Read an image from a file and superimpose some text over that image.

```python
img_logo=cv2.imread("/content/drive/MyDrive/Images_CV/logo.png",cv2.IMREAD_UNCHANGED)

position = (20,20)

cv2.putText(img_logo,"PYTHON PROGRAMMING",position,cv2.FONT_HERSHEY_DUPLEX,0.3,(0,0,255),1)
print(img_logo)
```

```python
cv2_imshow(img_logo)
```

```python
blur=cv2.blur(img_logo,(20,20))
cv2.imwrite("Blurred_img_logo.png",blur)
```

```python
cv2_imshow(blur)
```

```python
blurring=np.hstack((img_logo,blur))
cv2_imshow(blurring)
```

## Reading an image and displaying it using matplotlib

```python
img3 = cv2.imread("/content/drive/MyDrive/Images_CV/1.png")
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
plt.imshow(img3)
cv2.imwrite("Test_Write.png",img3)
```

## Read an image and convert it into grayscale and hsv formats.

```python
img4=cv2.imread("/content/drive/MyDrive/Images_CV/logo.png")
gray_scale=cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_scale)
cv2.imwrite("sample_gray.png",gray_scale)
```

```python
img5=cv2.imread("/content/drive/MyDrive/Images_CV/logo.png")
hsv_image=cv2.cvtColor(img5,cv2.COLOR_BGR2HSV)
plt.imshow(img5)
cv2.imwrite("HSV_image.png",hsv_image)
```

## Resizing an Image

```python
img6=cv2.imread("/content/drive/MyDrive/Images_CV/logo.png")
smaller=cv2.resize(img6,(100,100),interpolation=3)
plt.imshow(smaller)
cv2.imwrite("Resized_Image.png",smaller)
```
